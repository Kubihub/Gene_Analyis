```python
plt.figure(figsize = (5,5), dpi = 100)
sns.boxplot(data=data)
plt.tight_layout()
plt.xticks(rotation=90)
plt.title('Uncertainty checking on the dataset')
plt.savefig('entiredataset.jpg')
```

```python
sns.boxplot(data=data, x='SalePrice')
plt.title('Checking Uncertainty in SalePrice')
```

```python
sns.boxplot(data=data, y='SalePrice', x='SaleType')
plt.xlabel("SaleType")
plt.ylabel("SalePrice")
plt.title('Relationship between the SalePrice and Saletype on Houses')
plt.xticks(rotation=90)
```

```python
# Plotting the data with error bars
means = data.mean()
errors = data.std()

plt.errorbar(x=means.index, y= means, yerr=errors, fmt='o', capsize=5,)

# Adding labels and title
plt.xlabel('Data')
plt.ylabel('counts')
plt.title('Plot with Error Bars')
plt.xticks(rotation=90)

# Display the plot
plt.show()
```
